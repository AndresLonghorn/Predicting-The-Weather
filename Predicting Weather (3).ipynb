```python
from dash import Dash, html, dcc, Input, Output
import plotly.express as px


# Correct chronological order for months
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Sample dataset for demonstration
data = {
    "MONTH": ["January", "February", "March", "April", "May", "June", 
              "July", "August", "September", "October", "November", "December"],
    "BASEL_temp_mean": [30, 32, 35, 31, 33, 34, 36, 37, 38, 40, 41, 43],
    "BUDAPEST_temp_mean": [25, 27, 30, 26, 28, 29, 31, 32, 34, 35, 37, 39],
    "DE_BILT_temp_mean": [20, 22, 25, 21, 23, 24, 26, 27, 29, 30, 32, 34],
    "DRESDEN_temp_mean": [28, 29, 32, 27, 31, 30, 33, 34, 35, 36, 37, 38],
    "DUSSELDORF_temp_mean": [34, 36, 38, 33, 35, 37, 39, 40, 41, 43, 44, 45],
    "HEATHROW_temp_mean": [40, 42, 44, 41, 43, 45, 46, 47, 48, 50, 51, 52],
    "KASSEL_temp_mean": [20, 21, 23, 22, 24, 25, 26, 27, 28, 29, 30, 31],
    "LJUBLJANA_temp_mean": [23, 25, 28, 24, 26, 27, 29, 30, 32, 33, 34, 35],
    "MAASTRICHT_temp_mean": [27, 29, 31, 28, 30, 31, 33, 34, 35, 36, 38, 39],
    "MALMO_temp_mean": [25, 26, 28, 27, 29, 30, 31, 32, 34, 35, 37, 38],
    "MONTELIMAR_temp_mean": [32, 34, 36, 33, 35, 36, 37, 38, 40, 41, 43, 44],
    "MUENCHEN_temp_mean": [22, 23, 25, 24, 26, 27, 28, 29, 30, 31, 32, 33],
    "OSLO_temp_mean": [17, 19, 21, 18, 20, 22, 23, 24, 25, 27, 28, 29],
    "PERPIGNAN_temp_mean": [30, 32, 34, 31, 33, 34, 35, 36, 37, 38, 40, 41],
    "ROMA_temp_mean": [32, 34, 36, 33, 35, 37, 38, 39, 40, 42, 43, 44],
    "SONNBLICK_temp_mean": [18, 20, 22, 19, 21, 22, 24, 25, 26, 27, 28, 29],
    "STOCKHOLM_temp_mean": [19, 21, 23, 20, 22, 24, 25, 26, 28, 29, 30, 31],
    "TOURS_temp_mean": [26, 28, 30, 27, 29, 30, 31, 32, 33, 34, 36, 37],
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert MONTH to categorical with correct order
df["MONTH"] = pd.Categorical(df["MONTH"], categories=month_order, ordered=True)

# Group by MONTH and calculate mean for temperature columns
mean_temp = df.groupby('MONTH').mean()

# Remove "_temp_mean" from each column name
mean_temp.columns = mean_temp.columns.str.replace("_temp_mean", "", regex=False)

# Reset index to make MONTH a column again
mean_temp.reset_index(inplace=True)

# Melt the DataFrame for visualization (long format for px)
long_df = mean_temp.melt(id_vars=["MONTH"], var_name="capital", value_name="avg_temp")

# Initialize Dash app
app = Dash(__name__)

# App layout
app.layout = html.Div([
    html.H4('Average Temperatures by Month and Capital'),
    html.P("Select display option:"),
    dcc.RadioItems(
        id='selection',
        options=[
            {"label": "Monthly Scatter Plot", "value": "scatter"},
            {"label": "Monthly Bar Chart", "value": "bar"}
        ],
        value='scatter',
    ),
    dcc.Loading(dcc.Graph(id="graph"), type="cube")
])

# Callback to update the graph
@app.callback(
    Output("graph", "figure"), 
    Input("selection", "value"))
def update_graph(selection):
    if selection == "scatter":
        fig = px.scatter(
            long_df, x="MONTH", y="avg_temp", color="capital",
            title="Average Temperatures by Month and Capital (Scatter Plot)",
            labels={"avg_temp": "Average Temperature (°F)", "MONTH": "Month"},
            template="plotly",
        )
    elif selection == "bar":
        fig = px.bar(
            long_df, x="MONTH", y="avg_temp", color="capital",
            title="Average Temperatures of Capital (Bar Chart)",
            labels={"avg_temp": "Average Temperature (°F)", "MONTH": "Month"},
            template="plotly",
            barmode="group",
        )
    return fig

# Run server
if __name__ == "__main__":
    app.run_server(debug=True)
```
